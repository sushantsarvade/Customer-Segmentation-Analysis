```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\SUSHANT\\Documents\\Customer Segmentation Analysis\\data\\Customers.csv")
df.head()
```

```python
df.isnull().sum()

df.drop_duplicates(inplace=True)

df.rename(columns={'Annual_Income_(k$)': 'Annual_Income'}, inplace=True)

df.rename(columns={'Genre': 'Gender'}, inplace=True)
```

```python
df.describe()
```

```python
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
plt.hist(df['Age'], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(df['Annual_Income'], bins=10)
plt.title("Annual Income Distribution")
plt.show()
```

```python
plt.hist(df['Spending_Score'], bins=10)
plt.title("Spending Score Distribution")
plt.show()
```

```python
plt.scatter(df['Annual_Income'], df['Spending_Score'])
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Income vs Spending")
plt.show()
```

```python
plt.scatter(df['Age'], df['Spending_Score'])
plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.title("Age vs Spending")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```
